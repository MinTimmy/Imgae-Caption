# remove inf and nan

```python
def replace_nan(t):
  return tf.where(tf.math.is_nan(t), tf.zeros_like(t), t)
def replace_inf(t):
  return tf.where(tf.math.is_inf(t), tf.zeros_like(t), t)
```

# Train Dataset

```python

first = True
```

```python
dataset = tf.data.experimental.load(
      path = '/content/drive/MyDrive/Special_Topic/Top_Down_Bttom_Up/Code/Dataset/train_dataset_1', 
    )
first = False
```

```python
count = 0
# for num in range(0,30):
# for num in range(20, 40):
for num in range(30,45):
  pickle_file = pickle_dir_path + '/' + str(num) + ".pickle"
  print(count, pickle_file)
  count += 1

  with open(pickle_file, 'rb') as fp:
    data = pickle.load(fp)
    for i in ['mean_pooled_image_feature_1', 'mean_pooled_image_feature_2', 'text_embedding']:
        data[i] = replace_nan(data[i])
        data[i] = replace_inf(data[i])

    if first:
      dataset = tf.data.Dataset.from_tensor_slices((
              (
                  data['mean_pooled_image_feature_2'],
                  data['mean_pooled_image_feature_1'],
                  data['text_embedding']
              ),
              (
                  data['label'],
              ),
      )) 
      first = False
    else:
      tmp_dataset = tf.data.Dataset.from_tensor_slices((
              (
                  data['mean_pooled_image_feature_2'],
                  data['mean_pooled_image_feature_1'],
                  data['text_embedding']
               ),
              (
                data['label'],
              )
      )) 
      dataset = dataset.concatenate(tmp_dataset)
      del tmp_dataset
    del data
```

```python

print('saved', tf.data.experimental.cardinality(dataset).numpy())
```

```python
tf.data.experimental.save(
  dataset = dataset, 
  path = '/content/drive/MyDrive/Special_Topic/Top_Down_Bttom_Up/Code/Dataset/train_dataset', 
)
del dataset
```
