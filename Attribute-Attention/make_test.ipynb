# remove inf and nan

```python
def replace_nan(t):
  return tf.where(tf.math.is_nan(t), tf.zeros_like(t), t)
def replace_inf(t):
  return tf.where(tf.math.is_inf(t), tf.zeros_like(t), t)
```

```python
# for i in ['embedding_text', 'img', 'regions_set']:
for i in ['mean_pooled_image_feature_1', 'mean_pooled_image_feature_2', 'text_embedding']:
  dataset[i] = replace_nan(dataset[i])
  dataset[i] = replace_inf(dataset[i])
```
