```python
#         http://images.cocodataset.org/train2017/000000218439.jpg
# Example http://images.cocodataset.org/train2017/000000218439.jpg
def make_image_url(image_id, image_type = 'train'):
  image_id_string = str(image_id)
  while len(image_id_string) < 12:
    image_id_string = '0' + image_id_string
  return 'http://images.cocodataset.org/' + image_type + '2017/' + image_id_string + '.jpg'
```

```python
make_image_url(image_id, image_type = 'train')
```
