# remove inf and nan

```python
def replace_nan(t):
  return tf.where(tf.math.is_nan(t), tf.zeros_like(t), t)
def replace_inf(t):
  return tf.where(tf.math.is_inf(t), tf.zeros_like(t), t)
```

# Train Dataset

```python

first = True
```

```python
dataset = tf.data.experimental.load(
      path = '/content/drive/MyDrive/Special_Topic/Show_Control_Tell/Finial/Demo4/Dataset/train_dataset1', 
    )
first = False
```

```python

print('saved', tf.data.experimental.cardinality(dataset).numpy())
```

```python
print(first)
```

```python
count = 0
# for num in range(0,20):
# for num in range(20, 40):
for num in range(40,45):
  pickle_file = pickle_dir_path + '/' + str(num) + ".pickle"
  print(count, pickle_file)
  count += 1

  with open(pickle_file, 'rb') as fp:
    data = pickle.load(fp)
    batch_num = len(data['image_id'])
    zeros = tf.zeros(shape = (batch_num, 1024), )
    for i in ['embedding_text', 'img', 'regions_set',  'r_t']:
        data[i] = replace_nan(data[i])
        data[i] = replace_inf(data[i])

    if first:
      dataset = tf.data.Dataset.from_tensor_slices((
              (
                  data['embedding_text'],
                  data['img'],
                  data['r_t'],
                  zeros
              ),
              (
                  data['y_t'],
                  data['g_t']
              ),
      )) 
      first = False
    else:
      tmp_dataset = tf.data.Dataset.from_tensor_slices((
              (
                  data['embedding_text'],
                  data['img'],
                  data['r_t'],
                  zeros
               ),
              (
                data['y_t'],
                data['g_t'],
              )
      )) 
      dataset = dataset.concatenate(tmp_dataset)
      del tmp_dataset
    del data
  # break
```

```python

print('saved', tf.data.experimental.cardinality(dataset).numpy())
```

```python
tf.data.experimental.save(
  dataset = dataset, 
  path = '/content/drive/MyDrive/Special_Topic/Show_Control_Tell/Finial/Demo4/Dataset/train_dataset2', 
)
del dataset
```
